```python
def clashByTime(*courses):
    # maps day -> {time -> course}
    time_map = {}
    
    for course in courses:
        day, time = routine[course]  # assuming routine[course] = [day, time]
        
        if day not in time_map:
            time_map[day] = {}
        
        # Check if the same time already exists for this day
        if time in time_map[day]:
            # Clash found
            return course, time_map[day][time]
        
        # No clash, store this course
        time_map[day][time] = course
    
    # No clashes
    return []

# Example usage:
clashByTime("CSE 1112","CSE 3313","ENG 1011")
```
