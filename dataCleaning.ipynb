```python
import pandas as pd
```

```python
'''
collect the routine
convert to csv
convert to df
clean data to look like this:
{day1:  {t1:[eee2122, cse 2232,...]}
        {t2:[eee23122, cse 2232,...]}
    
}

'''
```

```python
raw_routine = pd.read_csv('raw_data.csv')
raw_routine.tail(10)
```

```python
raw_routine.drop(["Exam"], axis = 1, inplace = True)
raw_routine.tail()
```

```python
raw_routine.reset_index(drop=True)
```

```python
raw_routine['Day']=raw_routine['Day'].str.replace('Day',"")
raw_routine['Day']=raw_routine['Day'].astype(int)
```

```python
# type(raw_routine['Day'][0])
raw_routine
```

```python
raw_routine['Time']=raw_routine['Time'].str.replace('9:00 AM - 11:00 AM',"T1")
raw_routine['Time']=raw_routine['Time'].str.replace('11:30 AM - 1:30 PM',"T2")
raw_routine['Time']=raw_routine['Time'].str.replace('2:00 PM - 4:00 PM',"T3")
```

```python
raw_routine
```

```python
raw_routine['Courses']=raw_routine['Courses'].str.replace(r"\(.*?\)", "", regex=True).str.strip()
raw_routine
```

```python
# Keep only rows where 'Courses' does NOT start with "DS"
raw_routine = raw_routine[~raw_routine["Courses"].str.startswith("DS")]
raw_routine
```

```python
raw_routine['Courses']=raw_routine['Courses'].str.replace('/'," ")
raw_routine
```

```python
raw_routine["Courses"] = raw_routine["Courses"].str.extract(r"(\b[A-Za-z]+\s\d{4}\b)")

raw_routine
```

```python
raw_routine.dropna(inplace=True)
```

```python
new_dict = {row['Courses']: [row['Day'], row['Time']] for _, row in raw_routine.iterrows()}
raw_routine=pd.DataFrame(new_dict)
```
